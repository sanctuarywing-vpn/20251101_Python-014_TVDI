```python
import time

def download_file(file_name):
    print(f"開始下載 {file_name}...")
    time.sleep(2)  # 模擬下載時間
    print(f"{file_name} 下載完成！")

def main_sync():
    start_time  = time.time()
    download_file("檔案A.pdf")
    download_file("檔案B.pdf")
    download_file("檔案C.pdf")
    end_time = time.time()
    print(f"Elapsed time: {(end_time - start_time):.2f} 秒") # :.2f 表示保留兩位小數

if __name__ == "__main__":
    main_sync()

#程式一條一條執行，等待前一條執行完畢後才執行下一條
```

```python
import asyncio
import nest_asyncio

nest_asyncio.apply()

async def download_file(file_name):
    print(f"開始下載 {file_name}...")
    await asyncio.sleep(2)  # 模擬下載時間
    print(f"{file_name} 下載完成！")

async def main_sync():
    start_time  = time.time()
    await asyncio.gather(
        download_file("檔案A.pdf"),
        download_file("檔案B.pdf"),
        download_file("檔案C.pdf")
    )
    end_time = time.time()
    print(f"Elapsed time: {(end_time - start_time):.2f} 秒") # :.2f 表示保留兩位小數

if __name__ == "__main__":
    asyncio.run(main_sync())

#程式同時跑，有空就執行
```

async def	定義非同步函式	寫一張「工作清單」，還沒開始做。
await	等待任務完成	盯著微波爐直到東西熱好 (但 asyncio 允許你在這段時間去切菜)。
asyncio.gather	並行執行多個任務	一次把三個鍋子都開火，同時煮三道菜。
nest_asyncio	解決 Jupyter 報錯神器	讓你在夢中還能做夢 (允許事件迴圈嵌套)。

適合：I/O 密集型任務 (爬蟲、呼叫 API、讀寫資料庫)。
不適合：CPU 密集型任務 (影像處理、複雜數學運算) -> 請改用 multiprocessing。
