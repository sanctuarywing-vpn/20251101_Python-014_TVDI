數學運算子 + - * / // % **
+ 加法 5+8=13
- 減法 90-10=80
* 乘法 4*7=28
/ 浮點數除法 7/2=3.5 (小數點)
// 整數除法 7/2=3
% 餘數 7%3=1
** 次方 3**4=81

所有數學運算都是二元運算子
運算元 運算子 運算元 = 結果

```python
print(5+8)
print(90-10)
print(4*7)
print(7/2)
print(7//2)
print(7%3)
print(3**4)
print(99**99)
```
