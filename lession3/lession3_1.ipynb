```python
#字串和變數的整合
apple=5
orange=9
print("蘋果="+"個，橘子="+"個")
print("蘋果="+str(apple)+"個，橘子="+str(orange)+"個")
#使用str()將字串變成數值

apple=int(input("請輸入蘋果數量"))
orange=int(input("請輸入橘子數量"))
print("蘋果="+str(apple)+"個，橘子="+str(orange)+"個")
#若未使用 print() 主動顯示執行結果,則只會顯示最後一行
```

```python
#str template
#字串差補變數
#字串和變數整合，傳出1個全新的字串
print(f"蘋果={apple}個,橘子={orange}個")
#注意是 {} 而不是 ()
```
