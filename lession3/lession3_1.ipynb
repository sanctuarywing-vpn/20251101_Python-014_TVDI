什麼是 try-except 在 Python 裡，try-except 是一種錯誤處理機制，用來避免程式因為遇到錯誤而直接崩潰。它的用途是：當程式執行到某段可能會出錯的程式碼時，可以「嘗試 (try)」執行，如果真的發生錯誤，就「捕捉 (except)」並處理，而不是讓程式整個停掉。

1. try:
    # 嘗試執行這段程式碼
    risky_code = int(input("請輸入一個整數："))
    print("平方是：", risky_code ** 2)
except ValueError:
    # 如果發生錯誤 (例如輸入不是數字)，就執行這裡
    print("輸入錯誤，請輸入數字！")

2. try:
    f = open("data.txt", "r")
    content = f.read()
except FileNotFoundError:
    print("檔案不存在！")
finally:
    print("程式結束，無論有沒有錯誤都會執行這裡")
