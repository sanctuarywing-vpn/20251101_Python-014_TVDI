```python
nums = [1,2,3,4,5,6] #正取 0~5 ,負取 -6~-1
#透過索引取值
print(nums[0]) #抓取第一位值
print(nums[5])

#透過slice取值
print(nums[0:3]) #抓第1~第3位值
print(nums[:3]) #前面沒寫預設為最前開始抓到前3個
print(nums[-3:-1]) #負向開始抓
print(nums[-3:]) #後面沒寫則從第4個抓到最後

print("—————分隔線—————")

for num in nums: #依序列出
    print(num)
```

```python
print(type(nums)) #確認值的種類
```

```python
nums.append(7) #附加一個值附加一個值
nums
```

```python
nums.insert(0,-1) #在索引"0"處插入一個"-1"
nums
```

```python
nums.clear() #清除值
nums
```
