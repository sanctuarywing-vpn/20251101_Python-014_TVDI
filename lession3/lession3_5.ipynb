```python
nums = [1,2,3,4,5,6]
#透過索引取值
print(nums[0])
print(nums[5])

#透過slice取值
print(nums[0:3])
print(nums[:3]) #前面沒寫預設為最前開始抓到前3個
print(nums[-3:-1]) #第4第5
print(nums[-3:]) #後面沒寫則從第4個抓到最後

print("—————分隔線—————")

for num in nums: #依序列出
    print(num)
```
