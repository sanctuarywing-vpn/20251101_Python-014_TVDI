程式的流程控制: if 判斷, for 迴圈
if boolean(單項) + else boolean(雙向)

```python
age=17
if age>=18:
    print("滿18了")
    print("可以考駕照")
else:
    print("未滿18")
    print("不可以考駕照")
#if :  , else : "注意有冒號"
```

```python
#巢狀判斷
age=35
if age<18:
    print("未滿18")
    print("青少年")
else:
    if age<40:
        print("青年")
        print("每5年健康檢查")
    else:
        print("壯年")
        print("每3年健康檢查")

#多項選擇
#若else下還有if,可以整合成elif
age=65
if age<18:
    print("未滿18")
    print("青少年")
elif age<40:
    print("青年")
    print("每5年健康檢查")
elif age<60:
    print("壯年")
    print("每3年健康檢查")
else:
    print("老年")
    print("每年健康檢查")
```

```python
score = 98
if score>=90:
    print("優")
    print("你超棒")
elif score>=80:
    print("甲")
    print("很好")
elif score>=70:
    print("乙")
    print("繼續保持")
elif score>=60:
    print("丙")
    print("有進步空間")
else:
    print("丁")
    print("再加油")

#較程式邏輯寫法
```

```python
#可讀性較高，但效果一樣較繁多
score = 75
if score>=90:
    print("優")
elif 90 > score >= 80:
    print("甲")
elif score < 80 and score >= 70:
    print("乙")
elif score < 70 and score >= 60:
    print("丙")
else:
    print("丁")
```

```python
#random
import random
ran=random.randint(1,10) #在指定範圍內隨機數值，但不知道要幾次
if ran ==5:
    print("這是5")
else:
    print("這不是5")

print("—————分隔線—————")

#若要知道次數則加上while函數直到回圈到該值
while random.randint(1,10) !=5: #當i等於5 則跑外部迴圈
    print("這不是5")
print("這是5 (隨機到!=5 外部迴圈)")

print("—————分隔線—————")

#另一種寫法，迴圈到5後break
while True:
    rand_value=random.randint(1,10)
    if rand_value==5: #若值為5 則停止迴圈
        print("這是5(break 停止迴圈)")
        break
    print("這不是5")
```

- while True → 無限迴圈
- break → 強制結束迴圈
- continue → 跳過一次迴圈剩下的程式碼

while True:
    command = input("輸入指令 (exit 結束)：")
    if command == "exit":
        break   # 跳出迴圈
    print("你輸入了：", command)

for i in range(10):
    if i == 5:
        break   # 當 i=5 時跳出迴圈
    print(i)

for i in range(5):
    if i == 2:
        continue   # 當 i=2 時跳過，不執行 print
    print(i)
