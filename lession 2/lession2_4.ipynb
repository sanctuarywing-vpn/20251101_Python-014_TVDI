```python
# 複合指定運算子使用方法  #
# 參考官網python數學函式庫 https://docs.python.org/zh-tw/3/library/functions.html
# 外部函式庫 https://numpy.org/
# 科學家函式庫 https://scipy.org/

import math

chinese=78
chinese+=5  # 等同於 chinese=chinese+5
chinese*=10

#chinese 開平方根
value=math.sqrt(chinese)
value=round(value,ndigits=2)  # 四捨五入到小數點後兩位
#合併兩式
value=round(math.sqrt(chinese),ndigits=2)
value
```
